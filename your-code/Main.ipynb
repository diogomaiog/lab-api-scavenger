````python
#url = 'https://api.github.com/user'
#GET /repos/:owner/:repo/git/commits/:commit_sha

url = 'https://api.github.com/repos/ta-data-lis/lab-mysql-first-queries/commits'
response = requests.get(url,auth = HTTPBasicAuth(user, mypass))
print(f'Status Code: {response.status_code}')
commits = pd.DataFrame(response.json())
commits.head(10)
````

````python
pd.json_normalize(commits['commit'])
````

````python
# there are columns to unpack, let's use json_normalize
commits = pd.json_normalize(response.json())
commits.head()
````

````python
commits["commits_date"] = [datetime.strptime(timestamp, '%Y-%m-%dT%H:%M:%SZ') for timestamp in commits['commit.author.date']]
commits["commits_date"].head()
````

````python
def last_two_weeks(date):
    s_two_weeks = 60*60*24*14
    date = date.timestamp()
    now = datetime.now().timestamp()  
#    print(date)
#    print(now)
    if now - date > s_two_weeks:
        return 0
    else:
        return 1
    
commits["last_two_weeks"] = [last_two_weeks(date) for date in commits["commits_date"]]
cil_two_weeks = commits["last_two_weeks"].sum()
print(f'Commits in last two weeks: {cil_two_weeks}')
````
